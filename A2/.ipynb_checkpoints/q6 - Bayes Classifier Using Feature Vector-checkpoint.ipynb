```python
import math
from unittest import result
import pandas as pd
from re import M
import numpy as np
from sympy import *
import matplotlib.pyplot as plt

def equation(c1, Pw, MU, test):
    # u1 = np.mean(c1,axis=0)
    # cov_mat1 = np.stack(c1,axis = 1)
    # cov1 = np.cov(cov_mat1)
    inv1 = np.linalg.inv(c1)
    tra = np.transpose(np.array(test - MU))
    p1 = np.dot(tra,inv1)
    p2 = np.dot(p1,np.array(test - MU))
    equation = (-0.5*p2) - (math.log(2*np.pi)) + math.log(Pw) 
    return equation


def meanval(data):
    MU = []
    for i in range(128):
        s1 = 0.0
        for j in range(len(data)):
            s1 = s1 + data[j][i]
            
        MU1 = s1/395
        MU.append(MU1)
    MU = list(MU)    
    return MU




data = pd.read_csv("ffv.csv")
data = data.drop(['id'], axis=1)
data = data.drop(['result'], axis=1)

# #req data in dataframe format
male = data[5:400]
female = data[405:800]

testmale = data[0:5]
testfemale = data[400:405]


#covvariance matrix in dataframe format
malecov = male.cov()
femalecov = female.cov()

#covvariance in numpy format
mc = malecov.to_numpy()
fc = femalecov.to_numpy()

#inv-covvariance in numpy format
invcovmc = np.linalg.inv(mc)
invcovfc = np.linalg.inv(fc)

#req data in numpy format
male = male.to_numpy()
female = female.to_numpy()

testmale = testmale.to_numpy()
testfemale = testfemale.to_numpy()

print(mc)

X = [Symbol('x'+str(i)) for i in range(0, 128)] # 128 dimensional feature vector

Pw1 = 0.5
Pw2 = 0.5

MU1 = meanval(male)
MU2 = meanval(female)

m1=0
m2=0

print("Testing Male vectors")
for i in range(0,5):
    score1 = equation(mc,Pw1,MU1,testmale[i])
    score2 = equation(fc,Pw2,MU2,testmale[i])
    if(score2>score1):
        print("The vector ",i + 1," Belongs to Female")
        m1 = m1 + 1
    else:
       print("The vector ",i + 1," Belongs to Male")
print("Testing Female Vectors")
for i in range(0,5):
    score1 = equation(mc,Pw1,MU1,testfemale[i])
    score2 = equation(fc,Pw2,MU2,testfemale[i])
    if(score1>score2):
        print("The vector ",i + 401," Belongs to Male")
        m2 = m2 + 1
    else:
        print("The vector ",i + 401," Belongs to Female")
accuracy1 = (1 - m1/10)*100
accuracy2 = (1 - m2/10)*100
print("The Accuracy for Class 1: ",accuracy1,"%")
print("The Accuracy for Class 2: ",accuracy2,"%")
print("The Accuracy of Bayes Classifier: ",(accuracy1+accuracy2)/2,"%")




```
