```python
import cv2
from matplotlib import pyplot as plt
import numpy as np
import math

query = cv2.imread("queryimage.png", 1)
# converting BGR to RGB
query = cv2.cvtColor(query, cv2.COLOR_BGR2RGB)

ref1 = cv2.imread("refimage1.png", 1)
# converting BGR to RGB
ref1 = cv2.cvtColor(ref1, cv2.COLOR_BGR2RGB)

ref2 = cv2.imread("refimage2.png", 1)
# converting BGR to RGB
ref2 = cv2.cvtColor(ref2, cv2.COLOR_BGR2RGB)
```

```python
# Function to calculate frequency of each pixel
def frequency_pix(img):
    frequency = [0] * 257
    
    for i in img:
        for pix in i:
            mean = (int(pix[0]) + int(pix[1]) + int(pix[2]))//3
            frequency[mean] = frequency[mean] + 1
    return frequency
    

# User defined function to find histogram
def histogram(img):

    frequency = frequency_pix(img)
    intensity = []

    for i in range(257):
        intensity.append(i)

    graph = plt.plot(intensity, frequency, color = 'red')
    plt.title("Histogram using user-defined function")
    plt.show(graph)
    
    #in-built function for finding histogram
    hist1 = cv2.calcHist([img], [0], None, [256], [0,256])
    plt.plot(hist1)
    plt.title("Histogram using in-built function")
    plt.show()
    
    return
```

```python
# Display Histograms of each image
print("Query Image :")
freq_query = frequency_pix(query)
print(histogram(query))
```

```python
print("Reference Image 1 :")
freq_ref1 = frequency_pix(ref1)
print(histogram(ref1))
```

```python
print("Reference Image 2 :")
freq_ref2 = frequency_pix(ref2)
print(histogram(ref2))
```

```python
# FINDING COSINE DISTANCE

# Normalise Dimension Vector

# Function to normalise the frequency vector
def norm(freq):
    densq = 0
    norm = []
    for i in freq:
        densq = densq + (i * i)
    den = math.sqrt(densq)

    for i in freq:
        val = i / den
        norm.append(val)
    return norm

#Normalised Vectors
norm_q = norm(freq_query)
norm_ref1 = norm(freq_ref1)
norm_ref2 = norm(freq_ref2)
```

```python
# Cosine Distance

def cosine_dist(f1, f2):
    cos_val = 0
    for i in range(len(f1)):
        cos_val = cos_val + (f1[i] * f2[i])
        
    cosine_dist = 1 - cos_val
    return cosine_dist
```

```python
# Cosine Distance between Query Img and Reference 1
print("Cosine distance between Query & Reference 1 = ", cosine_dist(norm_q, norm_ref1))
print("Cosine distance between Query & Reference 2 = ", cosine_dist(norm_q, norm_ref2))
print("\nReference Image 1 is more similar to Query Image!!")
```
