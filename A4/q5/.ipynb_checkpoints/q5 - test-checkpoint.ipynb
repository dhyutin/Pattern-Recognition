```python
from cgi import test
import math
from turtle import shape
import cv2
import numpy as np
import pandas as pd



traindata = pd.read_csv("dataset_pca.csv")
testdata = pd.read_csv("dataset_test_pca.csv")

traindata = traindata.drop(traindata.columns[0], axis=1)
traindata = traindata.drop(traindata.columns[0], axis=1)


traindata = traindata.to_numpy()
testdata = testdata.to_numpy()

means = []
x = 0
for i in range(40):
    a = traindata[x]
    x+=1
    for j in range(8):
        c = np.add(a,traindata[x])
        x+=1
        a = c
    mean = c/9
    means.append(mean)
    print(mean)
    print(mean.shape)
means = np.array(means)

print(type(means))
print(means.shape)

for test in testdata:
    counter = 1 
    output_class = 0
    dis = 10000
    for ref in means:
        diff = np.subtract(ref,test)
        sum = 0
        for i in diff:
            sum = sum + (i*i)
        d = math.sqrt(sum)
        if(d<dis):
            dis = d
            output_class = counter
        counter+=1
    # print(test)
    print("resultant class = "+str(output_class))

```
