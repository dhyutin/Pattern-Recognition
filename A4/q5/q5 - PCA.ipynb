```python
import math
from turtle import shape
from unittest import result
import pandas as pd
from re import M
import numpy as np
from sympy import *
import matplotlib.pyplot as plt

def meanval(data):
    MU = []
    print("data shape")
    print(shape(data))
    print(shape(data)[1])

    for i in range(shape(data)[0]):
        s1 = 0.0
        for j in range(shape(data)[1]):
            s1 = s1 + data[i][j]
            
        MU1 = s1/shape(data)[1]
        MU.append(MU1)
    MU = list(MU)    
    return MU




data = pd.read_csv("dataset.csv")
data = data.drop(data.columns[0], axis=1)
data = data.drop(data.columns[0], axis=1)

td = pd.read_csv("dataset_test.csv")
td = td.drop(td.columns[0], axis=1)
td = td.drop(td.columns[0], axis=1)
print(data)



traindata = data.to_numpy()
testdata = td.to_numpy()


#PCA ---------------------------------
mu = []    
for i in range(256):
    m = 0
    for j in range(len(traindata)):
        m = m + traindata[j][i]
    m = m/360
    mu.append(m)
print(len(mu))

Z = []

for i in range(len(traindata)):
    x = np.subtract(traindata[i],mu)
    Z.append(x)
# print(len(Z))
Zt = np.transpose(Z)
# print(len(Zt))

COV = np.dot(Zt, Z)
COV = ((1/360)*COV)
# print(COV)
# print(type(COV))

w_, v_ = np.linalg.eig(COV)
w = w_ #eigen values
w_ = -np.sort(-w_) #eigen values descending order

# print(w)
# print(w_)
# print(v_)

# print(len(traindata[:,0]))

d_dash = 200
colval = []
for j in range(d_dash):
    p = np.where(w == w_[j])
    # print("index: ")
    # print(p[0][0])
    colval1 = traindata[:, p[0][0]]
    # print(shape(colval1))
    colval.append(colval1)

colval_test = []
for j in range(d_dash):
    p = np.where(w == w_[j])
    # print("index: ")
    # print(p[0][0])
    colval1 = testdata[:, p[0][0]]
    # print(shape(colval1))
    colval_test.append(colval1)

print(len(colval_test[0]))
print(type(colval_test))

values = np.array(colval)
values = values.transpose()

tval = np.array(colval_test)
tval = tval.transpose()

print(values.shape)

df = pd.DataFrame(values)
print(df)
df.to_csv("dataset_pca.csv")

df = pd.DataFrame(tval)
print(df)
df.to_csv("dataset_test_pca.csv")
```
