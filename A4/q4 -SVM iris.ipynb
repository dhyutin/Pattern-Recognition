```python
# Perceptron
import pandas as pd
from operator import add
import numpy as np
import matplotlib.pyplot as plt

data = pd.read_csv("Iris.csv")
data = data.drop(['SepalLengthCm', "PetalWidthCm", "Id"], axis = 1)
data = data.to_numpy()
for dat in data:
    if(dat[2] == "Iris-setosa"):
        dat[2] = 0
    elif(dat[2] == "Iris-versicolor"):
        dat[2] = 1
    else:
        dat[2] = 2
train = []
for i in range(len(data)):
    train.append([data[i][0], data[i][1], int(1), int(data[i][2])])


weight = [0,0,0]

checker = 0
i=0
iter_counter = 0

train = train[0:100]

while(checker!=100):
    prod = np.dot(train[i][0:3], weight)
    if((prod < 0)|(prod == 0)):
        checker = 0
        learntrain = np.dot(0.01,train[i][0:3])
        weight = list(map(add, weight, learntrain))
    else:
        checker+=1
    print("i = "+str(i))
    print(prod)
    print(weight)
    print("checker = "+str(checker))
    i+=1
    if(i==100):
        i=0
    iter_counter+=1
print("#Iterations = "+str(iter_counter))


# train = np.array(train)
print(data)
print(train)
```
