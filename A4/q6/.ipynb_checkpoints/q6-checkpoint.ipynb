```python
import math
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis as LDA
import pandas as pd
import numpy as np


data = pd.read_csv("dataset.csv")
y_train = data['result']
data = data.drop(['id'], axis=1)
X_train = data.drop(['result'], axis=1)

td = pd.read_csv("dataset_test.csv")
td = td.drop(['id'], axis=1)
X_test = td.drop(['result'], axis=1)
print(data)

lda = LDA(n_components=1)
X_train = lda.fit_transform(X_train, y_train)
X_test = lda.transform(X_test)

X_train = np.square(X_train)
X_test = np.square(X_test)

print(X_train)
print(X_test)
means = []
x = 0
for i in range(40):
    sum = 0
    for j in range(9):
        sum = sum + X_train[x]
        x+=1
    mean = sum/9
    means.append(mean)


for test in X_test:
    counter = 1 
    output_class = 0
    dis = 10000
    for ref in means:
        diff = np.subtract(ref,test)
        sum = 0
        for i in diff:
            sum = sum + (i*i)
        d = math.sqrt(sum)
        if(d<dis):
            dis = d
            output_class = counter
        counter+=1
    # print(test)
    print("resultant class = "+str(output_class))
   
```
