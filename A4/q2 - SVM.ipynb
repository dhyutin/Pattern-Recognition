```python
# Perceptron
import numpy as np
from operator import add
import matplotlib.pyplot as plt
```

```python
train = [[-2,-2,-1,1], [-1,-3,1,0], [1,-2,-1,1], [0,-1,1,0], [-1,-3,-1,1], [-1,-2,1,0], [1,-2,1,0], [1,1,-1,1]]
x = [2,-1,1,-1,-1,0,-1,1]
y = [2,2,3,-1,-3,-1,-2,-2]
weight = [0,0,0]
```

```python
checker = 0
i=0
iter_counter = 0

while(checker!=8):
    prod = np.dot(train[i][0:3], weight)
    if((prod < 0)|(prod == 0)):
        checker = 0
        learntrain = np.dot(0.01,train[i][0:3])
        weight = list(map(add, weight, learntrain))
    else:
        checker+=1
    print("i = "+str(i))
    print(prod)
    print(weight)
    print("checker = "+str(checker))
    i+=1
    if(i==8):
        i=0
    iter_counter+=1
print("#Iterations = "+str(iter_counter))
```

```python
plt.scatter(x[0:4], y[0:4], color= "green", s=30)
plt.scatter(x[4:8], y[4:8], color= "red", s=30)

plt.xlabel("X1")
plt.xlabel("X2")

x = np.array([-4, -3, -2, -1, 0, 1, 2, 3, 4])
y = -((weight[2] + (weight[0]*x))/weight[1])
plt.plot(x, y)
plt.show()
```

```python
# SVM 

train = [[-2,-2,-1,1], [-1,-3,1,0], [1,-2,-1,1], [0,-1,1,0], [-1,-3,-1,1], [-1,-2,1,0], [1,-2,1,0], [1,1,-1,1]]
x = [2,-1,1,-1,-1,0,-1,1]
y = [2,2,3,-1,-3,-1,-2,-2]
weight = [0,0,0]

checker = 0
i=0
iter_counter = 0


while(checker!=8):
    prod = np.dot(train[i][0:3], weight)
    if((prod < 1)|(prod == 1)):
        checker = 0
        learntrain = np.dot(0.5,train[i][0:3])
        weight = list(map(add, weight, learntrain))
    else:
        checker+=1
    print("i = "+str(i))
    print(prod)
    print(weight)
    print("checker = "+str(checker))
    i+=1
    if(i==8):
        i=0
    iter_counter+=1
print("#Iterations = "+str(iter_counter))
```
